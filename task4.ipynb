```python
#Implementing the mean(numbers) function :
def calculate_mean(numbers):
    
    mean= sum(numbers)/len(numbers)
    return mean
    

#Taking input from the user
numbers= input("Enter the numbers separated by space for mean calculation: ")
numbers_list= [int(i) for i in numbers.split()]

calculate_mean(numbers_list)

print(f"Mean of the given numbers is: {calculate_mean(numbers_list):.2f} ") 



   
```

```python
def calculate_mode(numbers):
    
    frequency = {}  # dictionary to store frequency of each number
    
    for num in numbers:
        
        if num in frequency:
            
            frequency[num] += 1
            
        else:
            
            frequency[num] = 1
            
    
    max_frequency = max(frequency.values())
    
    # if all numbers appear only once â†’ no mode
    if max_frequency == 1:
        
        return "No mode exists"
    
    mode_values = [num for num, freq in frequency.items() if freq == max_frequency]
    
    if len(mode_values) == 1:
        
        return mode_values[0]   # single mode
    
    else:
        
        return mode_values      # more than one mode

# Taking input from the user for mode calculation
numbers = input("Enter the numbers separated by space for mode calculation: ")
numbers_list = [int(i) for i in numbers.split()]

print("Mode of the given numbers is:", calculate_mode(numbers_list))
```

```python
#Implementing the median(numbers) function :

def calculate_median(numbers):
    # sorting the list
    numbers.sort()
    
    n = len(numbers)
    mid = n // 2   # middle index
    
    # Step 2: Odd number of elements
    if n % 2 != 0:
        return numbers[mid]
    else:
        # Step 3: Even number of elements â†’ average of two middle values
        return (numbers[mid - 1] + numbers[mid]) / 2


# Taking input from user
numbers = input("Enter the numbers separated by space for median calculation: ")
numbers_list = [int(i) for i in numbers.split()]

print("Median of the given numbers is:", calculate_median(numbers_list))
print(numbers_list)
```

```python
#Implementing the variance(numbers) function :
def calculate_variance(numbers):
    n = len(numbers)
    
    mean = sum(numbers) / n
    
    squared_diffs = [(x - mean) ** 2 for x in numbers]     #calculating squared differences
    
    variance = sum(squared_diffs) / n      # average of squared differences 
    
    return variance


# Taking input from user
numbers = input("Enter the numbers separated by space for variance calculation: ")
numbers_list = [int(i) for i in numbers.split()]

print("Variance of the given numbers is:", calculate_variance(numbers_list))

```

```python
#Implementing the std_dev(numbers) function :
import math

def std_dev(numbers):
    n = len(numbers)
    
    
    mean = sum(numbers) / n
    
    
    squared_diffs = [(x - mean) ** 2 for x in numbers]     #calculating squared differences
    
    
    variance = sum(squared_diffs) / n                   #average of squared differences 
    
   # standard deviation = square root of variance:
    return math.sqrt(variance)


#Taking input from user
numbers = input("Enter the numbers separated by space for standard deviation calculation: ")
numbers_list = [int(i) for i in numbers.split()]

print("Standard Deviation of the given numbers is:", std_dev(numbers_list))

```

```python
#Implementing the euclidean_distance(x, y) (for values & lists)  function :

import math

def euclidean_distance(x, y):
    # case 1: both are single numbers
       
    if isinstance(x, (int, float)) and isinstance(y, (int, float)):    #isinstance(obj,type), to check if obj is of this type
        return abs(x - y)       # absolute difference since distance never gonna be -ve
    
    # case 2: both are lists
    if isinstance(x, list) and isinstance(y, list):
        if len(x) != len(y):
            raise ValueError("Both lists must have the same length")
        
        squared_diffs = []
        for i in range(len(x)):
            squared_diffs.append((x[i] - y[i]) ** 2)
        
        return math.sqrt(sum(squared_diffs))
    
    raise TypeError("Inputs must be either both numbers or both lists")

# Taking input from user
choice = input("Calculate distance between (1) two numbers or (2) two lists? Enter 1 or 2: ")

if choice == '1':
    x = float(input("Enter first number: "))
    y = float(input("Enter second number: "))
    
    print("Euclidean distance between the two numbers is:", euclidean_distance(x, y))
    
    
elif choice == '2':
    x = input("Enter the first list of numbers separated by space: ")
    y = input("Enter the second list of numbers separated by space: ")
    
    x_list = [float(i) for i in x.split()]
    y_list = [float(i) for i in y.split()]
    
    print("Euclidean distance between the two lists is:", euclidean_distance(x_list, y_list))
    
else:
    print("Invalid choice. Please enter 1 or 2.")   


```

```python
#Implementing the sigmoid(x) function :

import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Take user input

x = float(input("Enter a number to calculate sigmoid: "))
print("Sigmoid of", x, "is:", sigmoid(x))
```
